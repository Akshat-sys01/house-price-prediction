```python
import pandas as pd

data = pd.read_csv(r"C:\Academics 25-26\4th Sem\My Work\Projects\house_price_prediction\data\house_data.csv")

# Check missing values
print(data.isnull().sum())

# Remove rows (if very few missing)
data = data.dropna()

# Fill with mean (most common for numeric data)
data['bedrooms'].fillna(data['bedrooms'].mean(), inplace=True)
data['bathrooms'].fillna(data['bathrooms'].mean(), inplace=True)

# Fill with median (better for skewed data)
data['bedrooms'].fillna(data['bedrooms'].median(), inplace=True)

# Remove irrelevant columns
data = data.drop(columns=['id'], errors='ignore')

# Separating Features (X) and Target (y)
X = data[['area', 'bedrooms', 'bathrooms', 'age']]
y = data['price']

print(X.head())
print(y.head())
```
